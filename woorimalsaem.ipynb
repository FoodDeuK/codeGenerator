```python
import os
import re

f = open('result.csv', encoding="utf8")

i = 0
while True:
  line = f.readline()
  if line != '\n':
    i = i+1
    print("%d번째 줄 :"%i + line)
  if i==10:
    break
f.close()
```

```python
# Skip-gram

from tqdm import tqdm
from konlpy.tag import Mecab
from gensim.models import Word2Vec

%cd Mecab-ko-for-Google-Colab

mecab = Mecab()

f = open('result.csv', encoding="utf8")
lines = f.read().splitlines()

result = []

for line in tqdm(lines):
  if line:
    result.append(mecab.morphs(line))

len(result)

model_sg = Word2Vec(result, size=100, window=5, min_count=0, workers=4, sg=1)

model_sg.save('urimalsaem_220522_mincount_0_sg.model')
model_sg.wv.save_word2vec_format('urimalsaem_220522_mincount_0_sg')
```

```python
# CBOW

from tqdm import tqdm
from konlpy.tag import Mecab
from gensim.models import Word2Vec

%cd Mecab-ko-for-Google-Colab

mecab = Mecab()

f = open('result.csv', encoding="utf8")
lines = f.read().splitlines()

result = []

for line in tqdm(lines):
  if line:
    result.append(mecab.morphs(line))

len(result)

model_cbow = Word2Vec(result, size=300, window=5, min_count=0, workers=4, sg=0)

model_cbow.save('urimalsaem_220516.model_mincount_1_cbow')
model_cbow.wv.save_word2vec_format('urimalsaem_220516_mincount_1_cbow')
```

```python
output_01 = model_sg.wv.similarity("대한민국", "한국")
output_02 = model_sg.wv.similarity("남자", "여자")

print(output_01)
print(output_02)

model_result1 = model_sg.wv.most_similar("한국")
print(model_result1)

model_result2 = model_sg.wv.most_similar("반도체")
print(model_result2)
```

```python
output_01 = model_cbow.wv.similarity("대한민국", "한국")
output_02 = model_cbow.wv.similarity("남자", "여자")

print(output_01)
print(output_02)

model_result1 = model_cbow.wv.most_similar("한국")
print(model_result1)

model_result2 = model_cbow.wv.most_similar("반도체")
print(model_result2)
```
