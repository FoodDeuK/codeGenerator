```python
import pandas as pd
from glob import glob
import re

def write(before):
    p = re.compile("[^0-9]")
    if before is not None:  
        before = before.replace("-","")
        before = before.replace("^","")
        after = before.replace(":","")
        after = "".join(p.findall(after))
        return after
```

```python
files = glob('전체 내려받기_우리말샘_xls_20220302/*.xls')
```

```python
print(files)
```

```python
dataframe = pd.DataFrame()
for file in files:
    data = pd.read_excel(file,sheet_name='Sheet0')
    dataframe = dataframe.append(data)
```

```python
new_data=dataframe[['어휘','품사','뜻풀이']]
```

```python
new_data
```

```python
idx_nonWordClass=new_data['품사']!='품사 없음'
```

```python
idx_nonWordClass
```

```python
real_data=new_data[idx_nonWordClass]
```

```python
real_data
```

```python
real_data['어휘'] = real_data['어휘'].apply(write)
```

```python
real_data
```

```python
real_data.drop_duplicates(['어휘'],keep='first')
```
