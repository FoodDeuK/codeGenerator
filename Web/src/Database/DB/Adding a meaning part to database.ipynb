```python
import sqlite3
import pandas as pd

con=sqlite3.connect("realdict_20220407.db")
cur=con.cursor()
cur.execute("SELECT * FROM wordlist")
rows=cur.fetchall()
cols=[column[0] for column in cur.description]
data_df=pd.DataFrame.from_records(data=rows,columns=cols)
```

```python
data_df
```

```python
result_data1=pd.read_csv("./220518(1).csv")
result_data1
```

```python
result_data2=pd.read_csv("./test_result746.csv")
result_data2
```

```python
word_lst=data_df['Word'].astype(str).tolist()
word_lst
```

```python
meaning_lst=[]
for word in word_lst:    
    try:
        idx=result_data1[result_data1['어휘']==word].index
        meaning=result_data1.loc[idx[0]][1]
        meaning_lst.append(meaning)
    except:
        idx=result_data2[result_data2['Word']==word].index
        meaning=result_data2.loc[idx[0]][1]
        meaning_lst.append(meaning)
```
